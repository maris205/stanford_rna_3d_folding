```python
import torch
import torch.nn as nn
from transformers import AutoModelForSequenceClassification, AutoTokenizer

# 1. 检查 GPU 可用性并设置设备
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# 2. 加载本地训练好的模型并移到 GPU
trained_model = AutoModelForSequenceClassification.from_pretrained("gene_eng_gpt2_v0_rna3d_ft").to(device)

# 3. 提取基础 GPT-2 模型
base_gpt2_model = trained_model.transformer

# 4. 定义自定义模型
class CustomGPT2Regression(nn.Module):
    def __init__(self, base_model):
        super().__init__()
        self.model = base_model
        self.regression_head = nn.Linear(self.model.config.hidden_size, 3, bias=False)

    def forward(self, input_ids, attention_mask):
        outputs = self.model(input_ids=input_ids, attention_mask=attention_mask)
        last_hidden_state = outputs.last_hidden_state  # (batch_size, seq_len, hidden_size)
        last_token_hidden = last_hidden_state[:, -1, :]  # 最后一个标记
        coordinates = self.regression_head(last_token_hidden)  # (batch_size, 3)
        return coordinates

# 5. 初始化自定义模型并移到 GPU
custom_model = CustomGPT2Regression(base_gpt2_model).to(device)

# 6. 加载回归头权重
if hasattr(trained_model, 'score'):
    custom_model.regression_head.load_state_dict(trained_model.score.state_dict())
elif hasattr(trained_model, 'classifier'):
    custom_model.regression_head.load_state_dict(trained_model.classifier.state_dict())

# 7. 设置为评估模式
custom_model.eval()

# 8. 加载分词器
tokenizer = AutoTokenizer.from_pretrained("gene_eng_gpt2_v0_rna3d_ft")

def get_rna_pos(seq):
    """
    获得ran序列最后一个残基的三维坐标预测
    """
    # 10. 分词和填充
    inputs = tokenizer(seq, truncation=True, padding='max_length', max_length=256, return_tensors="pt")
    
    # 11. 将输入数据移到 GPU
    inputs = {k: v.to(device) for k, v in inputs.items()}
    
    # 12. 在 GPU 上进行推理
    with torch.no_grad():
        predictions = custom_model(**inputs).squeeze(0).cpu().numpy()  # 推理后移回 CPU 并转为 NumPy

    return predictions
    


# 9. 测试序列
test_seq = "GGCGTAAGGA"
get_rna_pos(test_seq)
```

```python
def get_rna_all_pos(sequence):
    """
    获得rna每个残基的三维坐标预测
    """
    seq_pos_list_local = []
    for i in range(0, len(sequence)):
        seq = sequence[0:i+1].replace("U","T") #U-->T
        #如果seq长度大于1024则只要最后面的1024个字符
        if len(seq) > 1024:
            seq = seq[-1024:]
        pos = get_rna_pos(seq).tolist()
        #print(seq, pos)
        seq_pos_list_local.append(pos)
    return seq_pos_list_local

def get_csv_data(target_id, seq):
    """
    get csv result of seq
    """
    seq_pos_list = get_rna_all_pos(seq)
    data_list = []
    for index, (res, pos_list) in enumerate(zip(seq, seq_pos_list)):
        #print(index, res, pos_list)
        data = {
            "ID":target_id +"_" +str(index+1),
            "resname":res,
            "resid":str(index+1),
            "x_1":pos_list[0],
            "y_1":pos_list[1],
            "z_1":pos_list[2],
            "x_2":pos_list[0],
            "y_2":pos_list[1],
            "z_2":pos_list[2],
            "x_3":pos_list[0],
            "y_3":pos_list[1],
            "z_3":pos_list[2],
            "x_4":pos_list[0],
            "y_4":pos_list[1],
            "z_4":pos_list[2],
            "x_5":pos_list[0],
            "y_5":pos_list[1],
            "z_5":pos_list[2],
        }
        data_list.append(data)
    return data_list
```

```python
csv_data_list = []
for target_id,seq in zip(df["target_id"], df["sequence"]):
    data_list = get_csv_data(target_id, seq)
    csv_data_list.extend(data_list)
```

```python
# 将数据列表转换为 DataFrame
df = pd.DataFrame(csv_data_list)

# 写入 CSV 文件
df.to_csv("submission.csv", index=False)
```
